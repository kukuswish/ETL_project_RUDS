```python
#imports
import pandas as pd
from sqlalchemy import create_engine
import pymysql
import re
pymysql.install_as_MySQLdb()

engine = create_engine("mysql://root:password@localhost/etl_project")
conn = engine.connect()
```

```python
#get school csv into dataframe
school="../Assets/salaries-by-region.csv"

sdf=pd.read_csv(school)

sdf
```

```python
sdf.drop_duplicates()

sdf
```

```python
#clean the college names
c=1
for index, row in sdf.iterrows():
    schoolName = row['School Name']
    if(schoolName.find('(') > 1):
        line = re.sub(r" \(.+?\)", "", schoolName)
        row['School Name']=line
        
    schoolName = row['School Name']
    if(schoolName.find('-')>1):
        line = re.sub(r" - ", "--", schoolName)
        row['School Name']=line
    
sdf.head(300)
```

```python
#get college list table
clist = pd.read_sql_query("Select * from college_list", con=engine)
clist.head()
```

```python
#merge tables
merged = sdf.merge(clist, left_on="School Name", right_on="college_name", how='outer')

merged
```
