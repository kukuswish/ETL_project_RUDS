```python
#imports
import pandas as pd
from sqlalchemy import create_engine
import pymysql
import re
pymysql.install_as_MySQLdb()

engine = create_engine("mysql://root:password@localhost/etl_project")
conn = engine.connect()
```

```python
#get school csv into dataframe
school="../Assets/salaries-by-region.csv"

sdf=pd.read_csv(school)

sdf
```

```python
sdf.drop_duplicates()

sdf
```

```python
#clean the college names
c=1
for index, row in sdf.iterrows():
    schoolName = row['School Name']
    if(schoolName.find('(') > 1):
        line = re.sub(r" \(.+?\)", "", schoolName)
        row['School Name']=line
        
    schoolName = row['School Name']
    if(schoolName.find('-')>1):
        line = re.sub(r" - ", "--", schoolName)
        row['School Name']=line
        
    schoolName = row['School Name']
    if(schoolName.find(',')>1):
        line = re.sub(r"\, ", "--", schoolName)
        row['School Name']=line
    
    
sdf.head(300)
```

```python
#get college list table
clist = pd.read_sql_query("Select * from college_list", con=engine)
clist
```

```python
#merge tables
merged = sdf.merge(clist, left_on="School Name", right_on="college_name", how='left')

#merged['college_list_id']= merged['college_list_id'].astype('int')

merged.head()
```

```python
#convert list id to int
merged.fillna(value=0, inplace=True)

merged['college_list_id'] = merged['college_list_id'].astype('int')
merged.head()
```

```python
#convert state id to int
merged.fillna(value=0, inplace=True)

merged['state_id'] = merged['state_id'].astype('int')
merged.head()
```

```python
#drop columns
merged.drop(columns={'college_name','college_city','state_id','Region'}, inplace=True)
merged
```

```python
#convert to float
s=[]
mid=[]
ten=[]
tf=[]
sf=[]
nine=[]
for a, b in merged.iterrows():
    line = re.sub(r"\$", "", b['Starting Median Salary'])
    line = re.sub(r"\,", "", line)
    s.append(line)
    
merged['starting_median']=pd.Series(s).values
merged
```

```python
for a, b in merged.iterrows():   
    line = re.sub(r"\$", "", b['Mid-Career Median Salary'])
    line = re.sub(r"\,", "", line)
    mid.append(line)

merged['mid_career_median']=pd.Series(mid).values
merged
```

```python
def make_float(strObj):
    if(strObj != 0):
        cash_number = strObj[1:]
        return float(cash_number.split(",")[0])*1000+float(cash_number.split(",")[1])
    else:
        return(float(0))
```

```python
#rename columns
merged.rename(columns={'Mid-Career 10th Percentile Salary': 'perecent_ten',
                      'Mid-Career 25th Percentile Salary':'percent_tf',
                      'Mid-Career 75th Percentile Salary': 'percent_sf',
                      'Mid-Career 90th Percentile Salary': 'percent_nn'}, inplace=True)
merged.head()
```

```python
merged['percent_tf']= merged['percent_tf'].map(make_float)
```

```python

merged['percent_ten']= merged['perecent_ten'].map(make_float)
merged.head()
```

```python
merged['percent_sf']= merged['percent_sf'].map(make_float)
merged['percent_nn']= merged['percent_nn'].map(make_float)
```
