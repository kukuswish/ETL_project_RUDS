```python
#imports
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
```

```python
#get states csv into dataframe
states_csv="../Assets/states.csv"

states_df=pd.read_csv(states_csv)

states_df.head()
```

```python
# Create Engine and Pass in MySQL Connection
engine = create_engine("mysql://root:password@localhost/etl_project")
conn = engine.connect()
conn
```

```python
engine.table_names()
```

```python
#rename columns
states_df.rename(columns={"State":"state_name", 'Abbreviation': 'state_abbv'}, inplace=True)
states_df.head()
```

```python
#insert into table 'state'
states_df.to_sql(name="state",con=engine,if_exists='append', index=False)
```

```python
pd.read_sql_query("Select * from state", con=engine)
```
