```python
#imports
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()

engine = create_engine("mysql://root:password@localhost/etl_project")
conn = engine.connect()
```

```python
#get school json into dataframe
school="../Assets/schoolInfo.json"

sdf=pd.read_json(school)

sdf.head()
```

```python
#clean dataframe
sdf_clean= sdf[['displayName','city','state']].copy()

sdf_clean.head()
```

```python
#join
state_list = pd.read_sql_query("Select * from state", con=engine)

merged= sdf_clean.merge(state_list, left_on="state", right_on="state_abbv", how='left')

merged.head()
```

```python
#drop columns
merged.drop(columns={'state','state_name','state_abbv'}, inplace=True)
```

```python
#rename columns
merged.rename(columns={'displayName':'college_name','city':'college_city'}, inplace=True)
merged.head()
```

```python
#insert into database
merged.to_sql(name="college_list",con=engine,if_exists='append', index=False)
```
