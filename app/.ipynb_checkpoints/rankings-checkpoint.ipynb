```python
#imports
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()

engine = create_engine("mysql://root:password@localhost/etl_project")
conn = engine.connect()
```

```python
#get school json into dataframe
school="../Assets/schoolInfo.json"

sdf=pd.read_json(school)

sdf.head()
```

```python
#clean dataframe
sdf_clean= sdf[['displayName','overallRank']].copy()

sdf_clean.head()
```

```python
#join
state_list = pd.read_sql_query("Select * from college_list", con=engine)

merged= sdf_clean.merge(state_list, left_on="displayName", right_on="college_name", how='left')

merged.head()
```

```python
merged.drop(columns={'displayName','college_name','college_city','state_id'}, inplace=True)
```

```python
#rename columns 
merged.rename(columns={'overallRank':'ranking_number'}, inplace=True)
merged.head()
```

```python
#add year
merged['ranking_year']='2017'
    
merged.head()
```

```python
#insert into database
merged.to_sql(name="rankings",con=engine,if_exists='append', index=False)
```

```python
pd.read_sql_query("Select * from rankings", con=engine)
```
